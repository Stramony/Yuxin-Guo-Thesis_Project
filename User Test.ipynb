```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

# Loading Questionnaire Dataset

```python
df = pd.read_csv('data/user_text_data.csv')
df
```

```python
# Caculate the average score and standard deviation
mean_std_df = df.describe().loc[['mean', 'std']]
print("Mean score and standard deviation：")
print(mean_std_df)
```

```python
# Caculate the correlation matrix
correlation_matrix = df[['emotional_accuracy', 'scene_adaptability', 'tone', 'functional_acceptance']].corr()
print("Variable correlation matrix：")
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation matrix between different variables")
plt.show()
```
